```python
numerical_columns = data.select_dtypes(include=np.number).columns
print(numerical_columns)
```

NB:The above result shows that Age and Salary columns are the only numerical columns upon which otlier detection can be carried out.

```python
print("Outlier Detection using IQR Method:")
for col in numerical_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

    if not outliers.empty:
        print(f"\nColumn: {col}")
        print(f"  Q1: {Q1:.2f}")
        print(f"  Q3: {Q3:.2f}")
        print(f"  IQR: {IQR:.2f}")
        print(f"  Lower Bound: {lower_bound:.2f}")
        print(f"  Upper Bound: {upper_bound:.2f}")
        print(f"  Number of outliers: {len(outliers)}")
        display(outliers)
        print("\n")
    else:
        print(f"No outliers found in column: {col}")
```

The above result shows the rows and particular columns with outliers.

```python
# Remove outliers from the DataFrame
data.drop(outliers.index, inplace=True)

print(f"Shape of DataFrame after removing outliers: {data.shape}")
```

### **Re-checking for Outliers after Remova**l
Let's run the outlier detection again to confirm that the outliers have been successfully removed.

```python
print("Outlier Detection using IQR Method (after removal):")
for col in numerical_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_after_removal = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

    if not outliers_after_removal.empty:
        print(f"\nColumn: {col}")
        print(f"  Number of outliers: {len(outliers_after_removal)}")
        display(outliers_after_removal)
    else:
        print(f"No outliers found in column: {col} (after removal)")
```

**NB:** Outliers can be either removed or capped. I.e., involves limiting the extreme values in your data to a certain threshold. Instead of removing them, values above an upper limit are replaced with that upper limit, and values below a lower limit are replaced with that lower limit.



**Impact on Analysis:**

**Removing:** Can lead to a loss of data, which might reduce the sample size and potentially the statistical power or generalizability of your findings. It can also bias the sample if outliers are removed non-randomly.

**Capping:** Retains all observations, preventing data loss. However, it can distort the distribution of the variable, potentially reducing the variance and making the data appear less spread out than it truly is. It also introduces an artificial upper/lower bound.
